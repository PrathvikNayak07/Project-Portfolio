```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('tmdb_5000_movies.csv')

# Show first 5 rows
df.head()
```

```python
# Info: columns, datatypes
df.info()

# Check missing values
df.isnull().sum()

# Describe numerical columns
df.describe()
```

```python
# Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Drop rows with missing budget or vote_average
df = df.dropna(subset=['budget', 'vote_average'])

# Remove rows where budget or revenue is 0
df = df[df['budget'] != 0]
df = df[df['revenue'] != 0]

# Create 'year' column
df['year'] = df['release_date'].dt.year

# Create 'profit' column
df['profit'] = df['revenue'] - df['budget']

# Number of genres
import ast
df['genres_list'] = df['genres'].apply(lambda x: [d['name'] for d in ast.literal_eval(x)])
df['num_genres'] = df['genres_list'].apply(len)

# Show the cleaned data
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='budget', y='vote_average', alpha=0.5)
plt.title('Budget vs IMDb Rating')
plt.xlabel('Budget (in USD)')
plt.ylabel('IMDb Rating')
plt.xscale('log')  # budgets vary a lot, so log scale makes sense
plt.show()
```

**Budget vs IMDb Rating**

We observe that while movies with higher budgets tend to have slightly better ratings, there are many exceptions, showing that budget alone does not guarantee a better movie rating.


```python
plt.figure(figsize=(12,6))
avg_rating_per_year = df.groupby('year')['vote_average'].mean()
sns.lineplot(x=avg_rating_per_year.index, y=avg_rating_per_year.values)
plt.title('Average IMDb Rating Over Years')
plt.xlabel('Year')
plt.ylabel('Average IMDb Rating')
plt.show()
```

**Average IMDb Rating Over the Years**

Ratings have fluctuated over the years with some peaks, indicating changing audience preferences or industry trends.


```python
plt.figure(figsize=(8,5))
sns.countplot(x='num_genres', data=df)
plt.title('Number of Genres per Movie')
plt.xlabel('Number of Genres')
plt.ylabel('Number of Movies')
plt.show()
```

**Number of Genres per Movie**

Most movies fall within 1-3 genres, with very few movies spanning many genres.


```python
plt.figure(figsize=(8,5))
sns.boxplot(x='num_genres', y='vote_average', data=df)
plt.title('IMDb Rating by Number of Genres')
plt.xlabel('Number of Genres')
plt.ylabel('IMDb Rating')
plt.show()
```

**IMDb Rating by Number of Genres**

Movies with 1-2 genres tend to have higher median ratings compared to those with many genres, suggesting focused genre films might be better received.


```python
plt.figure(figsize=(10,6))
sns.histplot(df['profit'], bins=50, kde=True)
plt.title('Distribution of Movie Profits')
plt.xlabel('Profit (Revenue - Budget)')
plt.ylabel('Number of Movies')
plt.xlim(-1e8, 1e9)  # adjust limits for clearer view
plt.show()
```

**Profit Distribution**

Most movies either break even or lose money, while a small number of blockbusters generate huge profits.


```python
plt.figure(figsize=(10,8))
numeric_cols = ['budget', 'revenue', 'profit', 'popularity', 'vote_average', 'vote_count', 'num_genres']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Numerical Features')
plt.show()
```

**Correlation Analysis**

Strong correlations exist between budget, revenue, and profit as expected. Movie ratings have low correlation with budget and profit, suggesting other factors influence movie success.
